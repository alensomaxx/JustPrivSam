```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,11,1)
print ("crisp set X:", x)
```

```python
A = fuzz.trimf(x,[0,0,5])
B = fuzz.trimf(x,[5,10,15])

print ("fuzzy set A:", A)
print ("fuzzy set B:", B)

#Visualizing the fuzzy operations
plt.figure(figsize=(10,8))

#Plotting set A
plt.subplot(2,2,1)
plt.plot(x , A,label="Set A", color= 'blue')
plt.title('Fuzzy Set A')
plt.xlabel('x')
plt.ylabel('Membership Degree')

#Plotting set B

plt.subplot(2,2,2)
plt.plot(x , B,label="Set B", color= 'red')
plt.title('Fuzzy Set B')
plt.xlabel('x')
plt.ylabel('Membership Degree')

#Displaying the results
plt.tight_layout()
plt.show()
```

```python
A_not = 1-A
print("notA:", A_not)

#Fuzzy AND operation
A_and_B = np.fmin(A, B)
print("Intersection: ",A_and_B)

#Fuzzy OR operation
A_or_B= np.fmax(A,B)
print("Union:", A_or_B)

#Visualizing the fuzzy operations
plt.figure(figsize=(12,8))

#Plotting set NOT A
plt.subplot(2,2,3)
plt.plot(x , A_not,label="NOT A", color= 'green')
plt.title('Fuzzy NOT A')
plt.xlabel('x')
plt.ylabel('Membership Degree')

#Plotting FUZZY AND

plt.subplot(2,2,4)
plt.plot(x , A_and_B,label="A AND B", color= 'purple')
plt.title('Fuzzy AND (A and B)')
plt.xlabel('x')
plt.ylabel('Membership Degree')

```
